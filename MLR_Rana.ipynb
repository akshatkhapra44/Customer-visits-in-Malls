```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
dataset = pd.read_csv("Pedestrian_Counting_System_-_Monthly__counts_per_hour.csv_")
print(dataset)
```

Check null value in the column

```python
dataset.info()
```

Drop unneeded columns

```python
dataset= dataset.drop(['ID', 'Date_Time', 'Sensor_Name'], axis= 1)
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

Change the Day and Month to Numeric value to proccess data with the our model.

```python
dataset = dataset.replace({"Day":{"Monday": 1,"Tuesday": 2,"Wednesday": 3,"Thursday": 4,"Friday": 5,"Saturday": 6,"Sunday": 7}})
```

```python
dataset = dataset.replace({"Month":{"January": 1,"February": 2,"March": 3,"April": 4,"May": 5,"June": 6,"July": 7,"August": 8,"September": 9,"October": 10,"November": 11,"December": 12}})
```

Display the dataset after chaging day and month to numeric value

```python
# Display the dataset
dataset.head()
```

Set the dependent and target variable

```python
x = dataset[['Year', 'Month', 'Mdate', 'Day', 'Time', 'Sensor_ID']]
y = dataset['Hourly_Counts']
```

```python
#Split the data with training and test dataset
```

```python
#Splitting the data into train,test data 
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

Implementing the linear model



```python
#Fitting the Multiple Linear Regression model
```

```python
mlr = LinearRegression()  
mlr.fit(x_train, y_train)

```

```python
print("Intercept: ", mlr.intercept_)
print("Coefficients:")
list(zip(x, mlr.coef_))
```

#Prediction of test set

```python
y_pred_mlr= mlr.predict(x_test)
```

#Predicted values

```python
print("Prediction for test set: {}".format(y_pred_mlr))
```

#Actual value and the predicted value

```python
mlr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_mlr})
mlr_diff.head()
```

```python
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))
print('R squared: {:.2f}'.format(mlr.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```

```python
plt.plot(y_pred_mlr) #y_predictions
plt.plot(y_test) #y_predictions
```
